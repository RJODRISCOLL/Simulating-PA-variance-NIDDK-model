```python
# read dat
BW = pd.read_csv('BW.csv')
PA = pd.read_csv('PA.csv')

#define index
BW.set_index('time', inplace = True)
PA.set_index('time', inplace = True)

#get rid of annoying cols
BW = BW.iloc[:,0:1]
PA = PA.iloc[:,0:1]

#filter na 
BW = BW[BW.index.notna()]
PA = PA[PA.index.notna()]

# standard index and length
ind = BW.index
ro=len(BW)
```

# first we generate the weights 
### wlm, wg, wl

```python
# wlm datasets
wlm_weights_1 = pd.DataFrame(np.linspace(75,(75*1.02),ro) + np.random.normal(0,0.5, ro))
wlm_weights_2 = pd.DataFrame(np.linspace(75,(75*0.98), ro) + np.random.normal(0,1, ro))
wlm_weights_3 = pd.DataFrame(np.linspace(75,(75*1), ro) + np.random.normal(0,2, ro))
wlm_weights_1.index = ind
wlm_weights_2.index = ind
wlm_weights_3.index = ind
wlm_weights_1.to_csv('wlm_weights_1.csv')
wlm_weights_2.to_csv('wlm_weights_2.csv')
wlm_weights_3.to_csv('wlm_weights_3.csv')
```

```python
#wg datasets 
wg_weights_1 =  pd.DataFrame(np.linspace(75,(75*1.1),ro) + np.random.normal(0,0.5, ro))
wg_weights_2 =  pd.DataFrame(np.linspace(75,(75*1.2), ro) + np.random.normal(0,1, ro))
wg_weights_3 =  pd.DataFrame(np.linspace(75,(75*1.3), ro) + np.random.normal(0,2, ro))
wg_weights_1.index = ind
wg_weights_2.index = ind
wg_weights_3.index = ind
wg_weights_1.to_csv('wg_weights_1.csv')
wg_weights_2.to_csv('wg_weights_2.csv')
wg_weights_3.to_csv('wg_weights_3.csv')
```

```python
#wl datasets 
wl_weights_1 =  pd.DataFrame(np.linspace(75,(75*1.1),ro) + np.random.normal(0,0.5, ro))
wl_weights_2 =  pd.DataFrame(np.linspace(75,(75*1.2), ro) + np.random.normal(0,1, ro))
wl_weights_3 =  pd.DataFrame(np.linspace(75,(75*1.3), ro) + np.random.normal(0,2, ro))
wl_weights_1.index = ind
wl_weights_2.index = ind
wl_weights_3.index = ind
wl_weights_1.to_csv('wl_weights_1.csv')
wl_weights_2.to_csv('wl_weights_2.csv')
wl_weights_3.to_csv('wl_weights_3.csv')
```

```python
# random walks of PA 
num_sim = 999 # run 1000 
mc = [] # empty list for results
for _ in range(num_sim):
    PA = []
    init = np.random.normal(10, 1.5) # define an initial paee value
    PA.append(init) # starting value
    for i in range(ro-1):
        pa_var = PA[i] + np.random.normal(0, 0.4) # random walk with random normal 
        if pa_var < 0.1: # make sure not negative 
            pa_var = 0.1
        PA.append(pa_var) # add to sim 
    mc.append(pd.DataFrame(PA)) # add to end

# join and tidy
x=pd.concat(mc, axis = 1)
x.columns = range(1,1000)
x.index = ind
x.to_csv('PA.csv')
```

```python
# plot
fig = plt.figure()
fig.suptitle('Monte Carlo Simulation: PA')
plt.plot(x)
plt.xlabel('Week of study')
plt.ylabel('Price')
plt.figure(figsize=(20,10)) 
plt.show()
```
